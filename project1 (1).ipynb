```python
data = data.sort_values('x')

X = data[['x']]
y = data['y']

plt.scatter(X,y)
plt.plot(X,model.predict(X))
plt.show()
```

```python
####The scatter plot shows the actual data points.
##The straight line represents the best fit line obtained from Linear Regression.
##The model shows a positive relationship between X and y, as the line slopes upward.
##Some points lie above and below the line, indicating small prediction errors.
```

```python
model.score(X, y)
```
